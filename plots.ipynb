```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = ['Robert', 'DistillBERT', 'MiniLM', 'Logistic Regression', 'Neural Network']
baseline_accuracy = [0.9169, 0.9454, 0.4970, 0.5017, 0.7300]
ctpe_accuracy = [0.9800, 0.9827, 0.9923, 0.1466, 0.7031]
epochs = [3, 3, 3, None, None]  # Epochs are provided for only some models
```

```python
# Bar Plot: Comparison of Accuracy
x = np.arange(len(models))
width = 0.35

plt.figure(figsize=(10, 6))
plt.bar(x - width/2, baseline_accuracy, width, label='Baseline (Bi-Encoder)', alpha=0.7)
plt.bar(x + width/2, ctpe_accuracy, width, label='CTPE (Cross-Encoder)', alpha=0.7)

plt.xlabel('Models', fontsize=14)
plt.ylabel('Accuracy', fontsize=14)
plt.title('Comparison of Accuracy Across Models', fontsize=16)
plt.xticks(x, models, rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Line Plot: Accuracy Trend
plt.figure(figsize=(10, 6))
plt.plot(models, baseline_accuracy, marker='o', label='Baseline (Bi-Encoder)', linewidth=2)
plt.plot(models, ctpe_accuracy, marker='s', label='CTPE (Cross-Encoder)', linewidth=2)

plt.xlabel('Models', fontsize=14)
plt.ylabel('Accuracy', fontsize=14)
plt.title('Accuracy Trend Across Models', fontsize=16)
plt.legend()
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
